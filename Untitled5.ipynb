```python
"""
Physics-Informed Neural Network (PINN) for Pure Membrane Theory of Plates
==========================================================================

GOVERNING EQUATIONS (Plane-stress equilibrium):
  ∂Nx/∂x + ∂Nxy/∂y + px = 0       (x-equilibrium)
  ∂Nxy/∂x + ∂Ny/∂y  + py = 0       (y-equilibrium)

where the in-plane stress resultants are:
  Nx  = Et/(1-ν²) * (εx  + ν*εy)
  Ny  = Et/(1-ν²) * (εy  + ν*εx)
  Nxy = Et/(2(1+ν)) * γxy

and strains come from displacements u(x,y), v(x,y):
  εx  = ∂u/∂x
  εy  = ∂v/∂y
  γxy = ∂u/∂y + ∂v/∂x

Substituting → Navier-Lamé PDEs in u and v:
  C11*u_xx + C12*u_yy + C13*v_xy + px = 0
  C13*u_xy + C12*v_xx + C11*v_yy + py = 0

where:
  C11 = Et/(1-ν²)         (stretching stiffness)
  C12 = Et/(2(1+ν))       (shear stiffness)
  C13 = Et*ν/(1-ν²) + C12 (coupling)

Problem Setup:
  Domain  : x ∈ [0, a],  y ∈ [0, b]  (unit square)
  Loading : Uniform body load  px = p0, py = 0
            (simulates in-plane pressure, e.g. thermal or prestress)
  BCs     : Simply-supported membrane —
              u = 0  on x=0 and x=a   (no horizontal displacement on left/right)
              v = 0  on y=0 and y=b   (no vertical displacement on top/bottom)
              Nxy = 0 on all edges     (zero shear traction, free to slide)

Network:
  Two output heads — one for u(x,y), one for v(x,y)
  5 hidden layers × 50 neurons, tanh activation
  (2nd-order PDE needs less depth than 4th-order plate bending)

Validation:
  Manufactured solution (Method of Manufactured Solutions — MMS):
    u_exact = A * sin(πx/a) * sin(πy/b)   → satisfies u=0 on x=0,a approximately
    v_exact = B * sin(πx/a) * sin(πy/b)
  Body load px, py back-computed from these exact fields analytically.
  This gives a rigorous closed-form reference to compare against.

Training:
  Phase 1 — Adam warm-up  (1000 steps, small collocation set)
  Phase 2 — L-BFGS-B      (up to 10000 iterations, full collocation set)

Install:
    pip install tensorflow numpy matplotlib scipy pyDOE
"""

import time
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from pyDOE import lhs
from scipy.optimize import minimize


# ═════════════════════════════════════════════════════════════════════════════
# 0.  Material & Problem Parameters
# ═════════════════════════════════════════════════════════════════════════════
E    = 200e9       # Young's modulus (Pa)
nu   = 0.3         # Poisson's ratio
t    = 0.005       # Thickness (m)
a, b = 1.0, 1.0    # Plate dimensions (m)

# Membrane stiffness constants
C11  = E * t / (1.0 - nu**2)           # axial stiffness
C12  = E * t / (2.0 * (1.0 + nu))      # shear stiffness
C13  = E * t * nu / (1.0 - nu**2) + C12  # coupling term

print("=" * 60)
print("Pure Membrane PINN")
print("=" * 60)
print(f"  E   = {E:.3e} Pa")
print(f"  nu  = {nu}")
print(f"  t   = {t} m")
print(f"  C11 = {C11:.4e} N/m")
print(f"  C12 = {C12:.4e} N/m")
print(f"  C13 = {C13:.4e} N/m")
print("=" * 60)


# ═════════════════════════════════════════════════════════════════════════════
# 1.  Manufactured Solution (MMS) for Validation
# ═════════════════════════════════════════════════════════════════════════════
#
# We choose:
#   u_exact(x,y) = A * sin(πx/a) * cos(πy/b)
#   v_exact(x,y) = B * cos(πx/a) * sin(πy/b)
#
# These automatically satisfy:
#   u = 0 at x=0, x=a   (sin vanishes)
#   v = 0 at y=0, y=b   (sin vanishes)
#   Nxy = 0 on all edges (by symmetry of chosen modes)
#
# Back-compute body loads px, py by substituting into the PDEs:
#   px = -(C11*u_xx + C12*u_yy + C13*v_xy)
#   py = -(C13*u_xy + C12*v_xx + C11*v_yy)

A_mms = 1e-4   # amplitude of u (m) — realistic membrane displacement
B_mms = 5e-5   # amplitude of v (m)

px0 = np.pi**2 * (C11 * A_mms / a**2 + C12 * A_mms / b**2
                  + C13 * B_mms / (a * b))
py0 = np.pi**2 * (C13 * A_mms / (a * b) + C12 * B_mms / a**2
                  + C11 * B_mms / b**2)

print(f"\nMMS body loads:")
print(f"  px = {px0:.4e} N/m²")
print(f"  py = {py0:.4e} N/m²")


def mms_exact(X, Y):
    """Returns exact u, v fields from MMS."""
    u = A_mms * np.sin(np.pi * X / a) * np.cos(np.pi * Y / b)
    v = B_mms * np.cos(np.pi * X / a) * np.sin(np.pi * Y / b)
    return u, v


# ═════════════════════════════════════════════════════════════════════════════
# 2.  Neural Network Architecture  (dual-output: predicts u and v together)
# ═════════════════════════════════════════════════════════════════════════════
class MembraneNet(tf.keras.Model):
    """
    Input  : (x, y)   — spatial coordinates
    Output : (u, v)   — in-plane displacements

    5 hidden layers × 50 neurons, tanh activation.
    tanh is essential — we need smooth 2nd-order derivatives.
    Dual output (2 neurons in final layer) predicts both fields simultaneously.
    This is more efficient than two separate networks.
    """

    def __init__(self, n_hidden=5, n_neurons=50):
        super().__init__()
        self.hidden = []
        for _ in range(n_hidden):
            self.hidden.append(
                tf.keras.layers.Dense(
                    n_neurons,
                    activation="tanh",
                    kernel_initializer="glorot_normal",
                    bias_initializer="zeros",
                )
            )
        # 2 outputs: u and v
        self.out = tf.keras.layers.Dense(
            2,
            kernel_initializer="glorot_normal",
            bias_initializer="zeros",
        )

    def call(self, x, y):
        z = tf.concat([x, y], axis=1)   # (N, 2)
        for layer in self.hidden:
            z = layer(z)
        out = self.out(z)               # (N, 2)
        u = out[:, 0:1]                 # (N, 1) — x-displacement
        v = out[:, 1:2]                 # (N, 1) — y-displacement
        return u, v


# ═════════════════════════════════════════════════════════════════════════════
# 3.  PDE Residual  (2-level tape — 2nd order PDE needs only 2 levels)
# ═════════════════════════════════════════════════════════════════════════════
def membrane_residual(model, x_f, y_f, px, py):
    """
    Computes the Navier-Lamé residuals:
      R1 = C11*u_xx + C12*u_yy + C13*v_xy + px
      R2 = C13*u_xy + C12*v_xx + C11*v_yy + py

    NOTE: Only 2 levels of GradientTape needed (vs 4 for biharmonic plate).
    This makes membrane PINN significantly cheaper to train per iteration.

    Returns R1, R2 — both should be zero at interior collocation points.
    """
    with tf.GradientTape(persistent=True) as t2:
        t2.watch([x_f, y_f])
        with tf.GradientTape(persistent=True) as t1:
            t1.watch([x_f, y_f])
            u, v = model(x_f, y_f)
        # First derivatives
        u_x = t1.gradient(u, x_f)
        u_y = t1.gradient(u, y_f)
        v_x = t1.gradient(v, x_f)
        v_y = t1.gradient(v, y_f)
        del t1

    # Second derivatives
    u_xx = t2.gradient(u_x, x_f)
    u_yy = t2.gradient(u_y, y_f)
    u_xy = t2.gradient(u_x, y_f)   # = t2.gradient(u_y, x_f) by symmetry

    v_xx = t2.gradient(v_x, x_f)
    v_yy = t2.gradient(v_y, y_f)
    v_xy = t2.gradient(v_x, y_f)
    del t2

    # Navier-Lamé PDEs
    R1 = C11 * u_xx + C12 * u_yy + C13 * v_xy + px
    R2 = C13 * u_xy + C12 * v_xx + C11 * v_yy + py

    return R1, R2


# ═════════════════════════════════════════════════════════════════════════════
# 4.  Stress Resultants (post-processing)
# ═════════════════════════════════════════════════════════════════════════════
def compute_stress_resultants(model, x_p, y_p):
    """
    Computes Nx, Ny, Nxy at given points from predicted u, v fields.
    Used for post-processing and validation only — not in the loss.
    """
    with tf.GradientTape(persistent=True) as t1:
        t1.watch([x_p, y_p])
        u, v = model(x_p, y_p)
    eps_x = t1.gradient(u, x_p)          # ∂u/∂x
    eps_y = t1.gradient(v, y_p)          # ∂v/∂y
    gam   = t1.gradient(u, y_p) + t1.gradient(v, x_p)  # ∂u/∂y + ∂v/∂x
    del t1

    Nx  = C11 * eps_x + C11 * nu * eps_y       # simplified for plane stress
    Ny  = C11 * eps_y + C11 * nu * eps_x
    Nxy = C12 * gam
    return Nx, Ny, Nxy


# ═════════════════════════════════════════════════════════════════════════════
# 5.  Loss Function
# ═════════════════════════════════════════════════════════════════════════════
def compute_loss(model,
                 x_f, y_f, px_tf, py_tf,
                 x_x0, y_x0,
                 x_xa, y_xa,
                 x_y0, y_y0,
                 x_yb, y_yb,
                 lam_bc=20.0):
    """
    Total Loss = PDE residual loss + BC loss

    PDE loss (interior, Nf points):
      R1² + R2² → both equilibrium equations must be satisfied

    BC loss (boundary, Nb points per edge):
      Simply-supported membrane:
        u = 0  on x=0 (left)  and x=a (right)
        v = 0  on y=0 (bottom) and y=b (top)
      These are enforced as soft constraints (penalized in loss).

    Note: Nxy=0 (zero shear traction) is a natural BC for this formulation
    and does NOT need to be explicitly enforced — it is automatically
    satisfied when u and v satisfy the equilibrium PDEs with zero shear
    loading on the boundary.

    lam_bc: weight for BC loss relative to PDE loss.
            Higher value → BCs satisfied more strictly at cost of PDE accuracy.
            Typical range: 10-100.
    """

    # ── PDE residual loss ─────────────────────────────────────────────────
    R1, R2 = membrane_residual(model, x_f, y_f, px_tf, py_tf)
    mse_pde = tf.reduce_mean(tf.square(R1)) + tf.reduce_mean(tf.square(R2))

    # ── BC loss: u=0 on left (x=0) and right (x=a) edges ─────────────────
    u_x0, _ = model(x_x0, y_x0)
    u_xa, _ = model(x_xa, y_xa)

    # ── BC loss: v=0 on bottom (y=0) and top (y=b) edges ─────────────────
    _, v_y0 = model(x_y0, y_y0)
    _, v_yb = model(x_yb, y_yb)

    mse_bc = (tf.reduce_mean(tf.square(u_x0)) +
              tf.reduce_mean(tf.square(u_xa)) +
              tf.reduce_mean(tf.square(v_y0)) +
              tf.reduce_mean(tf.square(v_yb)))

    total = mse_pde + lam_bc * mse_bc
    return total, mse_pde, mse_bc


# ═════════════════════════════════════════════════════════════════════════════
# 6.  JIT-Compiled Training Step
# ═════════════════════════════════════════════════════════════════════════════
@tf.function
def train_step(model, optimizer,
               x_f, y_f, px_tf, py_tf,
               x_x0, y_x0,
               x_xa, y_xa,
               x_y0, y_y0,
               x_yb, y_yb):
    with tf.GradientTape() as tape:
        loss, mse_pde, mse_bc = compute_loss(
            model,
            x_f, y_f, px_tf, py_tf,
            x_x0, y_x0,
            x_xa, y_xa,
            x_y0, y_y0,
            x_yb, y_yb,
        )
    grads = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(grads, model.trainable_variables))
    return loss, mse_pde, mse_bc


# ═════════════════════════════════════════════════════════════════════════════
# 7.  Training Data Generation
# ═════════════════════════════════════════════════════════════════════════════
def generate_data(Nf=8000, Nb=150):
    """
    Nf : interior collocation points (LHS sampling — better coverage than random)
    Nb : boundary points per edge
    """
    lb = np.array([0.0, 0.0])
    ub = np.array([a,   b  ])
    X_f  = lb + (ub - lb) * lhs(2, Nf)

    s = np.random.uniform
    X_x0 = np.hstack([np.zeros((Nb,1)),    s(0, b, (Nb,1))])
    X_xa = np.hstack([a*np.ones((Nb,1)),   s(0, b, (Nb,1))])
    X_y0 = np.hstack([s(0, a, (Nb,1)),     np.zeros((Nb,1))])
    X_yb = np.hstack([s(0, a, (Nb,1)),     b*np.ones((Nb,1))])
    return X_f, X_x0, X_xa, X_y0, X_yb


def to_tf(arr):
    return tf.constant(arr, dtype=tf.float32)


# ═════════════════════════════════════════════════════════════════════════════
# 8.  L-BFGS Trainer
# ═════════════════════════════════════════════════════════════════════════════
class LBFGSTrainer:
    def __init__(self, model, x_f, y_f, px_tf, py_tf,
                 x_x0, y_x0, x_xa, y_xa,
                 x_y0, y_y0, x_yb, y_yb):
        self.model  = model
        self.data   = (x_f, y_f, px_tf, py_tf,
                       x_x0, y_x0, x_xa, y_xa,
                       x_y0, y_y0, x_yb, y_yb)
        self.iter   = 0
        self.hist   = []
        self.shapes = [v.shape for v in model.trainable_variables]
        n = sum(int(np.prod(s)) for s in self.shapes)
        print(f"  Trainable parameters: {n}")

    def _pack(self):
        return np.concatenate(
            [v.numpy().flatten() for v in self.model.trainable_variables]
        )

    def _unpack(self, flat):
        off = 0
        for var, sh in zip(self.model.trainable_variables, self.shapes):
            sz = int(np.prod(sh))
            var.assign(flat[off:off+sz].reshape(sh))
            off += sz

    def loss_and_grad(self, flat):
        self._unpack(flat)
        x_f, y_f, px_tf, py_tf, \
        x_x0, y_x0, x_xa, y_xa, \
        x_y0, y_y0, x_yb, y_yb = self.data

        with tf.GradientTape() as tape:
            loss, mse_pde, mse_bc = compute_loss(
                self.model,
                x_f, y_f, px_tf, py_tf,
                x_x0, y_x0, x_xa, y_xa,
                x_y0, y_y0, x_yb, y_yb,
            )
        grads    = tape.gradient(loss, self.model.trainable_variables)
        flat_g   = np.concatenate([g.numpy().flatten() for g in grads])
        loss_val = float(loss.numpy())
        self.iter += 1

        if self.iter % 500 == 0:
            print(f"  Iter {self.iter:5d} | Loss {loss_val:.3e} | "
                  f"PDE {float(mse_pde):.3e} | BC {float(mse_bc):.3e}")
            self.hist.append(loss_val)

        return loss_val, flat_g.astype(np.float64)

    def train(self, max_iter=10_000):
        print("\nL-BFGS-B training ...")
        x0 = self._pack().astype(np.float64)
        res = minimize(self.loss_and_grad, x0, method="L-BFGS-B", jac=True,
                       options={"maxiter": max_iter, "maxfun": max_iter,
                                "ftol": 1e-12, "gtol": 1e-8, "disp": False})
        self._unpack(res.x)
        print(f"  Done — {res.message}")
        print(f"  Final loss: {res.fun:.3e}\n")
        return self.hist


# ═════════════════════════════════════════════════════════════════════════════
# 9.  Evaluation & Plotting
# ═════════════════════════════════════════════════════════════════════════════
def evaluate_and_plot(model, loss_hist):
    N = 100
    xv = np.linspace(0, a, N)
    yv = np.linspace(0, b, N)
    X, Y = np.meshgrid(xv, yv)

    x_flat = to_tf(X.flatten()[:,None])
    y_flat = to_tf(Y.flatten()[:,None])

    u_pred, v_pred = model(x_flat, y_flat)
    u_pred = u_pred.numpy().reshape(N, N)
    v_pred = v_pred.numpy().reshape(N, N)

    u_exact, v_exact = mms_exact(X, Y)

    # Errors
    rel_u = np.linalg.norm(u_pred - u_exact) / np.linalg.norm(u_exact)
    rel_v = np.linalg.norm(v_pred - v_exact) / np.linalg.norm(v_exact)
    print(f"Relative L² error  u : {rel_u:.3e}")
    print(f"Relative L² error  v : {rel_v:.3e}")

    # Stress resultants
    Nx, Ny, Nxy = compute_stress_resultants(model, x_flat, y_flat)
    Nx  = Nx.numpy().reshape(N, N)
    Ny  = Ny.numpy().reshape(N, N)
    Nxy = Nxy.numpy().reshape(N, N)

    # ── Figure 1: Displacement fields ────────────────────────────────────
    fig, axes = plt.subplots(2, 3, figsize=(16, 10))
    fig.suptitle(
        "PINN — Pure Membrane Theory\n"
        f"Rel. L² error:  u = {rel_u:.2e}   |   v = {rel_v:.2e}",
        fontsize=13
    )

    def _plot(ax, data, title, cmap="viridis", unit="m"):
        im = ax.contourf(X, Y, data, levels=40, cmap=cmap)
        plt.colorbar(im, ax=ax, label=unit)
        ax.set_title(title)
        ax.set_xlabel("x (m)")
        ax.set_ylabel("y (m)")

    _plot(axes[0,0], u_pred,   "PINN  u(x,y)")
    _plot(axes[0,1], u_exact,  "Exact u(x,y)")
    _plot(axes[0,2], np.abs(u_pred - u_exact), "|error| u", cmap="Reds")

    _plot(axes[1,0], v_pred,   "PINN  v(x,y)")
    _plot(axes[1,1], v_exact,  "Exact v(x,y)")
    _plot(axes[1,2], np.abs(v_pred - v_exact), "|error| v", cmap="Reds")

    plt.tight_layout()
    plt.savefig("/content/pinn_membrane_displacements.png", dpi=150, bbox_inches="tight")
    plt.show()
    print("Saved → /content/pinn_membrane_displacements.png")

    # ── Figure 2: Stress resultants ───────────────────────────────────────
    fig2, axes2 = plt.subplots(1, 3, figsize=(16, 4))
    fig2.suptitle("PINN — In-Plane Stress Resultants  Nx, Ny, Nxy", fontsize=13)
    _plot(axes2[0], Nx,  "Nx (N/m)",  unit="N/m")
    _plot(axes2[1], Ny,  "Ny (N/m)",  unit="N/m")
    _plot(axes2[2], Nxy, "Nxy (N/m)", unit="N/m")
    plt.tight_layout()
    plt.savefig("/content/pinn_membrane_stresses.png", dpi=150, bbox_inches="tight")
    plt.show()
    print("Saved → /content/pinn_membrane_stresses.png")

    # ── Figure 3: Centreline slices ───────────────────────────────────────
    mid = N // 2
    fig3, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
    ax1.plot(xv, u_exact[mid,:], "b-",  lw=2,   label="Exact")
    ax1.plot(xv, u_pred[mid,:],  "r--", lw=1.5, label="PINN")
    ax1.set_title(f"u(x, y={b/2:.2f})")
    ax1.set_xlabel("x (m)"); ax1.set_ylabel("u (m)")
    ax1.legend(); ax1.grid(alpha=0.3)

    ax2.plot(yv, v_exact[:,mid], "b-",  lw=2,   label="Exact")
    ax2.plot(yv, v_pred[:,mid],  "r--", lw=1.5, label="PINN")
    ax2.set_title(f"v(x={a/2:.2f}, y)")
    ax2.set_xlabel("y (m)"); ax2.set_ylabel("v (m)")
    ax2.legend(); ax2.grid(alpha=0.3)

    plt.tight_layout()
    plt.savefig("/content/pinn_membrane_slices.png", dpi=150, bbox_inches="tight")
    plt.show()
    print("Saved → /content/pinn_membrane_slices.png")

    # ── Figure 4: Training loss ────────────────────────────────────────────
    if loss_hist:
        fig4, ax = plt.subplots(figsize=(7, 4))
        ax.semilogy([i*500 for i in range(1, len(loss_hist)+1)],
                    loss_hist, "b-o", ms=4)
        ax.set_xlabel("Iteration"); ax.set_ylabel("Total Loss")
        ax.set_title("L-BFGS-B Training Loss")
        ax.grid(alpha=0.3)
        plt.savefig("/content/pinn_membrane_loss.png", dpi=150, bbox_inches="tight")
        plt.show()
        print("Saved → /content/pinn_membrane_loss.png")


# ═════════════════════════════════════════════════════════════════════════════
# 10.  Main
# ═════════════════════════════════════════════════════════════════════════════
if __name__ == "__main__":

    # tf.constants for body loads (avoids graph retracing)
    px_tf = tf.constant(px0, dtype=tf.float32)
    py_tf = tf.constant(py0, dtype=tf.float32)

    # Build model
    model = MembraneNet(n_hidden=5, n_neurons=50)

    # ── Phase 1: Adam warm-up on small dataset ────────────────────────────
    print("\nGenerating warm-up data (small) ...")
    X_f_w, X_x0_w, X_xa_w, X_y0_w, X_yb_w = generate_data(Nf=800, Nb=40)

    def split(arr):
        return to_tf(arr[:,0:1]), to_tf(arr[:,1:2])

    x_f_w, y_f_w   = split(X_f_w)
    x_x0_w,y_x0_w  = split(X_x0_w)
    x_xa_w,y_xa_w  = split(X_xa_w)
    x_y0_w,y_y0_w  = split(X_y0_w)
    x_yb_w,y_yb_w  = split(X_yb_w)

    optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3)

    # Verify JIT compilation timing
    print("\nVerifying @tf.function compilation:")
    for step in range(5):
        t0 = time.time()
        loss, mpde, mbc = train_step(
            model, optimizer,
            x_f_w, y_f_w, px_tf, py_tf,
            x_x0_w, y_x0_w, x_xa_w, y_xa_w,
            x_y0_w, y_y0_w, x_yb_w, y_yb_w,
        )
        print(f"  Step {step}: {time.time()-t0:.2f}s | Loss {float(loss):.3e}")
    print("  ✓ Steps 1-4 should be <0.3s if compilation worked.\n")

    print("Adam warm-up (1500 steps) ...")
    t0 = time.time()
    for step in range(1500):
        loss, mpde, mbc = train_step(
            model, optimizer,
            x_f_w, y_f_w, px_tf, py_tf,
            x_x0_w, y_x0_w, x_xa_w, y_xa_w,
            x_y0_w, y_y0_w, x_yb_w, y_yb_w,
        )
        if (step + 1) % 300 == 0:
            print(f"  Adam {step+1:4d} | Loss {float(loss):.3e} | "
                  f"PDE {float(mpde):.3e} | BC {float(mbc):.3e}")
    print(f"  Warm-up done in {time.time()-t0:.1f}s\n")

    # ── Phase 2: L-BFGS on full dataset ──────────────────────────────────
    print("Generating full training data ...")
    X_f, X_x0, X_xa, X_y0, X_yb = generate_data(Nf=8000, Nb=150)

    x_f, y_f     = split(X_f)
    x_x0, y_x0  = split(X_x0)
    x_xa, y_xa   = split(X_xa)
    x_y0, y_y0   = split(X_y0)
    x_yb, y_yb   = split(X_yb)

    trainer = LBFGSTrainer(
        model,
        x_f, y_f, px_tf, py_tf,
        x_x0, y_x0, x_xa, y_xa,
        x_y0, y_y0, x_yb, y_yb,
    )
    loss_hist = trainer.train(max_iter=10_000)

    # ── Evaluate & plot ───────────────────────────────────────────────────
    evaluate_and_plot(model, loss_hist)
```
